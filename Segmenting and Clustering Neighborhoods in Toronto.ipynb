```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Using beautiful soup and request I will webscrape this wikipedia page and find the table to sort through

```python
C_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(C_url).text
```

```python
soup = BeautifulSoup(source, 'xml')
```

```python
table=soup.find('table')
```

The dataframe will consist of 3 columns:  PostalCode, Borough, and Neighborhood

```python
column_names = ['Postalcode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)
```

```python
df
```

This code goes through each row and stores it

```python
for t_row in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data
```

```python
df.head(30)
```

Data cleaning will consist of deleting rows where a "Borough"is not assigned

```python
df=df[df['Borough']!='Not assigned']
```

```python
df.head(50)
```

This step will combine neighborhoods in the same postal code

```python
quick_df=df.groupby('Postalcode')['Neighborhood'].apply(lambda x: "%s" % ', '.join(x))
quick_df=quick_df.reset_index(drop=False)
quick_df.rename(columns={'Neighborhood':'Neighborhood_joined'},inplace=True)
```

```python
df_merge = pd.merge(df, quick_df, on='Postalcode')
```

```python
df_merge.drop(['Neighborhood'],axis=1,inplace=True)
```

```python
df_merge.rename(columns={'Neighborhood_joined':'Neighborhood'},inplace=True)
```

```python
df_merge.head()
```

```python
df_merge.drop_duplicates(inplace=True)
```
